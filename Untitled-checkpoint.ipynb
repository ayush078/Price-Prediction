```python
avocado_prophet_df = avocado_df[['Date', 'AveragePrice']] # here i only take date and averageprice to work on and made new data frame with prophet
```

```python
avocado_prophet_df
```

```python
avocado_prophet_df = avocado_prophet_df.rename(columns = {'Date': 'ds', 'AveragePrice': 'y'})
```

```python
avocado_prophet_df
```

```python
m = Prophet()
```

```python
m.fit(avocado_prophet_df) #train
```

```python
#forecast the future
future = m.make_future_dataframe(periods = 400)
forecast = m.predict(future)
```

```python
forecast
```

```python
figure = m.plot(forecast , xlabel = 'Date', ylabel= 'Price')
```

```python
figure = m.plot_components(forecast)
```
